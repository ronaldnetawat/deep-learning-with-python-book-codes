# 5.2 Evaluating machine learning models

- You can only control what you observe.
- Essential to be able to realiably measure the generalization power of our model.
- Introduce the different ways to evaluate ML models.

## 5.2.1 Training, validation, and test sets

- Spilitting data into training, validation, and test sets.
- Test data should be as similar as possible to the production data, then you can deploy your model in production.
- ***Hyperparameters***: # of layers, # of neurons etc.
- ***Parameters***: weights of the network.
- Tuning the config. can result in *overfitting to the validation set*.
- ***Information leaks***: everytime we tune the hyperparams based on the model's performance of the validation set, some info. about the val. set leaks into the model.
- For never-seen before data, we need the test set.
  - Model should have *NO* access to the test set.


When little data is available, 3 ways:
- ***Simple holdout validation***:
  - Simplest evaluation protocol.
  - Flaw: if too few samples, then validation and test set may contain too less samples to accurately represent the data at hand.
  - How to know: if different random shuffling of data before splitting give very different performance scores, they we have this issue.

- ***K-fold validation***:
  - Split data into K partitions of equal size.
  - For each partition i, train on K - 1 partitions, and evaluate it on partition i.
  - Final score is the average of K scores obtained.

- ***Iterated k-fold validation***:
  - Used when little data, and need to evaluate model as precisely as possible.
  - K-fold validation multiple times, shuffling data everytime before splitting it K ways.
  - Final score: average of scores obtained at each run of K-fold validation.
  - Expensive computationally.

## 5.2.2 Beating a common-sense baseline

- Besides evaluation protocols, use common-sense baselines.
- Before working on a dataset, pick a trivial baseline that you'll try to beat.
- Example: for MNIST classification, a simple baseline would be a validation accuracy > random classification (which is 0.1).
- Common-sense baseline is essential when solving a problem *no one has solved before*.


## 5.2.3 Things to keep in mind about model evaluation

- ***Data representativeness***: both training and test set should be representative of the data. Should *randomly shuffle* data before splitting.
- ***Arrow of time***: If predicting future given the past, DO NO SHUFFLE. This will create a *temporal leak*. Make sure all data is test is *posterior* to training data.
- ***Redundancy***: Make sure training and testing set are disjoint.

By evaluating the performance of model is how we can monitor the real most important tension at the center of ML: between *optimization & generalization*, between *underfitting & overfitting*.
